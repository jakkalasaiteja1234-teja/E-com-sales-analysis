```python
# step3: creating function to find the expries
def findexp():
    cnt=0
    for date in df['CC Exp Date']:
        if date.split('-')[1]=='20':
            cnt=cnt+1
        return cnt
findexp()       
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jakka\\OneDrive\\Documents\\EcomPur.csv')
#-------------------------------------------
#1.Purchase Price Analysis
#-------------------------------------------
#max,min,mean
max_price=df['Purchase Price'].max()
min_price=df['Purchase Price'].min()
mean_price=df['Purchase Price'].mean()
print("max price:",max_price)
print("min price:",min_price)
print("mean price:",mean_price)
#Plot Purchase price Distribution
plt.figure()
plt.hist(df['Purchase Price'],bins=10)
plt.title("purchase price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
#create folder if not exists
os.makedirs("images",exist_ok=True)
plt.savefig("images/Purchase_Price_distribution.png",dpi=300)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jakka\\OneDrive\\Documents\\EcomPur.csv')
#---------------------------
#2.Language Analysis(French)
#---------------------------
#count people who speak French
french_count=df[df['Language']=='fr'].shape[0]
print("People who speak French:",french_count)
#bar chart of language
plt.figure()
df['Language'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Language Share")
plt.ylabel("")
plt.savefig("images/language_pie.png",dpi=300)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jakka\\OneDrive\\Documents\\EcomPur.csv')
#------------------------------
#3.Job Analysis
#------------------------------
engineers=df[df['Job'].str.contains('Engineer',case=False)]
print("Number of Engineers:",engineers.shape[0])
#job distribution
plt.figure()
df['Job'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Jobs")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.savefig("images/language_pie.png",dpi=300)
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jakka\\OneDrive\\Documents\\EcomPur.csv')
plt.figure(figsize=(8,5))

plt.hist(
    df['Purchase Price'],
    bins=15,
    color='#FF6B6B',
    edgecolor='black'
)

plt.title("Purchase Price Distribution", fontsize=16, fontweight='bold')
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.savefig("images/price_distribution_beautiful.png", dpi=300)
plt.show()
```
